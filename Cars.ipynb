```python
def colors():
    import numpy as np
    return "#"+"".join(np.random.choice(list('abcdef123456789'), 6))
```

```python
index = cars['year'].value_counts().index
value = cars['year'].value_counts().values

label = {"fontsize":15, "color":"grey", "fontfamily":"serif"}
title = {"fontsize":15, "color":"#6acbff", "fontfamily":"serif"}

fig, ax = plt.subplots(nrows=1, ncols=1, facecolor='black', figsize=(17, 8), dpi=100, subplot_kw=dict(facecolor="black"))

ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

ax.set_xlabel("Year", fontdict=label)
ax.set_ylabel("Count", fontdict=label)

plt.bar(index, value, color=[colors() for c in range(len(index))])
plt.tick_params(labelcolor="white",color='white')
plt.xticks(rotation=0)

for p, val in zip(ax.patches, value):
    x, y = p.get_x()+0.1 , p.get_height()+0.2
    plt.text(x, y, val, color="white")

plt.show()
```
